```python
import numpy as np
import matplotlib.pyplot as plt

gNa = 120.0
gK = 40.0
gL = 0.3
Cm = 1.0
V_K = -80.0
V_Na= 50.0
V_L = -75.0

V = []
n = []
h = []
m = []
times = []
times.append(0)
n.append(0)
m.append(0)
h.append(1)
V.append(-75)

time = 0
timeFinal = 200
dt = 20/1000
```

#Tudo ou nada

```python
for i in range(10):
  n.clear()
  m.clear()
  h.clear()
  V.clear()
  times.clear()

  times.append(0)
  n.append(0)
  m.append(0)
  h.append(1)
  V.append(-75)
  time = 0
  Iapp = i

  while(time < timeFinal):
    alpha_m = 0.1*(V[-1]+40.0)/(1.0-np.exp(-(V[-1]+40.0)/10.0))
    beta_m  = 4.0*np.exp(-(V[-1]+65.0)/18.0)
    alpha_h = 0.07*np.exp(-(V[-1]+65.0)/20.0)
    beta_h  = 1.0/(1.0+np.exp(-(V[-1]+35.0)/10.0))
    alpha_n = 0.01*(V[-1]+55.0)/(1.0-np.exp(-(V[-1]+55.0)/10.0))
    beta_n  = 0.125*np.exp(-(V[-1]+65)/80.0)

    n.append(dt*(alpha_n*(1-n[-1]) - beta_n*n[-1]) + n[-1])
    m.append(dt*(alpha_m*(1-m[-1]) - beta_m*m[-1]) + m[-1])
    h.append(dt*(alpha_h*(1-h[-1]) - beta_h*h[-1]) + h[-1])
    V.append(V[-1] - (dt/Cm)*(gK*(n[-1]**4)*(V[-1] - V_K) + gNa*(m[-1]**3)*h[-1]*(V[-1] - V_Na) + gL*(V[-1] - V_L)) + (Iapp * ((1 + i*10 < time < 1.1 + i*10) )))

    time += dt
    times.append(time)

  plt.plot(times, V, label=("Iapp = " + str(i)))
  plt.legend()

plt.show()

```
