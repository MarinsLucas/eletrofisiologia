#Parâmetros


```python
import matplotlib.pyplot as plt
```

#Condições Iniciais

```python
M=1
k1 = (10**9)/M
km1 = 7.5 *(10**5)
k2 = 10e3

#COndição inicial
E0 = 10**(-5)*M
S0 = 100*E0
C0 = 0
P0=0
```

#Exercício 1: intervalo de tempo 0 até 10e-5

```python
S = S0
C = C0
P = P0
E = E0
time = 0
Time = 10**(-5)
dt = Time/100.0

times = []
concentraS = []
concentraC = []
concentraE = []
concentraP = []
while(time < Time):
  times.append(time)
  S = (-k1*S*E0 + k1*S*C + km1*C)*dt + S
  concentraS.append(S/S0)
  C = dt*(k1*E0*S -(k2 + km1 + k1*S)*C) + C
  concentraC.append(C/E0)
  E = E0-C
  concentraE.append(E/E0)
  P= S0-S-C
  concentraP.append(P/S0)
  time+= dt

plt.plot(times, concentraS, label="S")
plt.plot(times, concentraC, label="C")
plt.plot(times, concentraE, label="E")
plt.plot(times, concentraP, label="P")
plt.legend()

plt.show()
```
