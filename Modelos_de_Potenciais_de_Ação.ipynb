# Reduced Hodgkin-Huxley

```python
import numpy as np
import matplotlib.pyplot as plt

gNa = 120.0
gK = 40.0
gL = 0.3
Cm = 1.0
V_K = -80.0
V_Na= 50.0
V_L = -75.0

V = []
n = []
h = []
m = []
times = []
times.append(0)
n.append(0)
m.append(0)
h.append(0.8)
V.append(-75)

time = 0
timeFinal = 20
dt = 1/100
```

```python
n.clear()
m.clear()
h.clear()
V.clear()
times.clear()

times.append(0)
n.append(0)
m.append(0.00034080100065184073)
h.append(0.8)
V.append(-75)
time = 0
Iapp = 1

while(time < timeFinal):
  alpha_m = 0.1*(V[-1]+40.0)/(1.0-np.exp(-(V[-1]+40.0)/10.0))
  beta_m  = 4.0*np.exp(-(V[-1]+65.0)/18.0)
  alpha_h = 0.07*np.exp(-(V[-1]+65.0)/20.0)
  beta_h  = 1.0/(1.0+np.exp(-(V[-1]+35.0)/10.0))
  alpha_n = 0.01*(V[-1]+55.0)/(1.0-np.exp(-(V[-1]+55.0)/10.0))
  beta_n  = 0.125*np.exp(-(V[-1]+65)/80.0)

  h.append(dt*(alpha_h*(1-h[-1]) - beta_h*h[-1]) + h[-1])
  m_inf = alpha_m/(alpha_m+beta_m)
  n_inf = 0.8 - h[-1]

  V.append(V[-1] - (dt/Cm)*(gK*(n_inf**4)*(V[-1] - V_K) + gNa*(m_inf**3)*h[-1]*(V[-1] - V_Na) + gL*(V[-1] - V_L)) + (Iapp * ((1 < time < 2))))

  time += dt
  times.append(time)

plt.plot(times, V)
plt.title("Hodgkin-Huxley reduzido")
plt.xlabel("Tempo (ms)")
plt.ylabel("V (mV)")
plt.grid(True)
plt.show()


plt.plot(h, V)
plt.xlabel("h")
plt.ylabel("V (mV)")
plt.title("Plano de Fase HH reduzido")
plt.grid(True)
plt.show()
```

# FitzHugh-Nagumo

```python
x = [] #Potncial da membrana
y = [] #Resistncia do sistema

a = 0.8
b = 0.7
c = 5 #Varivel auxiliar

z  = 0.8 #Estmulo externo

time = 0
times = []
timeFinal = 30

dt = 2/10000
```

```python
x.clear()
y.clear()
times.clear()
time = 0
#Condies iniciais
times.append(0)

#A partir dos valores das constantes, foi observado que os valores abaixos eram os valores de equilbrio do sistema
x.append(-1.2276401612211987)
y.append(-0.610914516031346)

while(time < timeFinal):
  y.append((x[-1] + a - b*y[-1])*(dt/c) + y[-1])
  x.append(((((-x[-1])**3)/3) + x[-1] - y[-1] + z*(1  < time < 3))*dt + x[-1])

  time+= dt/2
  times.append(time)

plt.plot(times, x)
plt.grid(True)
plt.show()
plt.grid(True)
plt.plot(y,x)
plt.show()
```

Observando o comportamento do potncial de ao e do FitzHugh-Nagumo e o Hodgkin-Huxley reduzido,  possvel observar que existem semelhanas e diferenas. A semelhana  o comportamento generalizado, tanto em relao as tendncias de retorno a um ponto de equilbrio quanto o formato do plano de fase formando uma figura fechada (a partir do momento que h o estmulo, que seja suficiente para despolarizar, o potncial de ao ocorre por completo).

A diferena encontrada no exerccio,  devido a adequao dos parmetros dos dois modelos, para que haja uma maior correspondncia em relao as caractersticas do potncial de ao. Ou seja,  possvel observar que com os parmetros escolhidos, o modelo FHN apresentou fases de polarizao e despolarizao mais amenos, enquanto o modelo HH possu essas fases mais abruptas.

#Questo 2 - Mitchell-Schaeffer


```python
vgate = 0.13
tauopen = 120 #ms
tauclose = 180 #ms
tout =  6  #ms
tin = 0.3 #ms
vstim = 0.1 #ms
Cm = 10**-2
beta = 6.41025641*(10**5) #Razo volume superfcie

dt = 1/1000
gatefactor = []
V = []
```

```python
gatefactor.clear()
V.clear()
times.clear()
times.append(0)

gatefactor.append(1.0)
V.append(0.0)

time = 0
timeFinal = 800

while(time < timeFinal):
  Istim = 0.1 * ( 3 <time < 3.1)
  gatefactor.append(dt*((V[-1]<vgate)*((1-gatefactor[-1])/tauopen) + (V[-1]>vgate)*(-gatefactor[-1]/tauclose)) + gatefactor[-1])
  V.append((dt/Cm)*((gatefactor[-1]*(V[-1]**2)*(1-V[-1])/tin) - (V[-1]/tout) + Istim) + V[-1])
  time+=dt
  times.append(time)


plt.plot(times, V)
plt.title("Mitchell-Schaeffer")
plt.show()

plt.plot(gatefactor, V)
plt.title("Plano de fase MS")
plt.show()

```
