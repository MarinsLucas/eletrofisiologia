## Equation Description

The GHK voltage equation is given as:

$ V = \left(\frac{RT}{zF}\right) \ln \left(\frac{P_{\text{K}}[K^+]_{\text{out}} + P_{\text{Na}}[Na^+]_{\text{out}} + P_{\text{Cl}}[Cl^-]_{\text{in}}}{P_{\text{K}}[K^+]_{\text{in}} + P_{\text{Na}}[Na^+]_{\text{in}} + P_{\text{Cl}}[Cl^-]_{\text{out}}}\right)$

Where:
- \( V \) = Membrane potential
- \( R \) = Gas constant
- \( T \) = Absolute temperature
- \( z \) = Ion charge
- \( F \) = Faraday constant
- \( P \) = Permeability of ions
- \( [X]_{\text{in}} \) = Intracellular concentration of ion X
- \( [X]_{\text{out}} \) = Extracellular concentration of ion X

```python
import math
import matplotlib.pyplot as plt

# Function to calculate the membrane potential (V) using GHK equation
def calculate_membrane_potential(P_K, P_Na, K_out, Na_out, K_in, Na_in):
    # Constants
    R = 8.314  # Gas constant (J/(mol*K))
    T = 310    # Temperature (K)
    z = 1      # Ion charge (e.g., +1 for K+ or Na+)
    F = 96485  # Faraday constant (C/mol)

    # GHK equation
    numerator = (P_K * K_out + P_Na * Na_out)
    denominator = (P_K * K_in + P_Na * Na_in)
    V = (R * T / (z * F)) * math.log(numerator / denominator)

    return V

# Function to create Gaussian curve for changing permeabilities over time
def gaussian_function(x, mean, stdev, scale):
    exponent = -((x - mean) ** 2) / (2 * stdev ** 2)
    return scale * math.exp(exponent)

# Initial concentrations and parameters
K_out = 5   # Extracellular Concentration of Potassium (mM)
Na_out = 150  # Extracellular Concentration of Sodium (mM)
K_in = 140  # Intracellular Concentration of Potassium (mM)
Na_in = 15  # Intracellular Concentration of Sodium (mM)
time_steps = 200  # Number of time steps
P_Na = 0.01
P_K = 0.5

stime = 10


# Lists to store time and voltage values
time = []
voltage = []
pk = []
pna = []
pulse = []
DV = 0
# Simulating changes in ion permeabilities over time using Gaussian functions
for i in range(time_steps):


    # Calculate membrane potential at each time step
    V = calculate_membrane_potential(P_K, P_Na, K_out, Na_out, K_in, Na_in)
    # Using a Gaussian function for sodium and potassium permeability within the specified voltage range

    P_Na = gaussian_function(i, mean=1+stime, stdev=5, scale=1) + 0.01
    P_K =  gaussian_function(i, mean=40+stime, stdev=10, scale=1) + 0.5

    # Append time and voltage values to the lists
    time.append(i)
    voltage.append(V)
    pk.append(P_K)
    pna.append(P_Na)

# Plotting the voltage vs time
plt.figure(figsize=(8, 6))
plt.plot(time, voltage, color='green')
plt.title('Membrane Potential vs Time')
plt.xlabel('Time')
plt.ylabel('Membrane Potential (V)')
plt.grid(True)
plt.show()
# Plotting the voltage vs time
plt.figure(figsize=(8, 6))
plt.plot(time, pk, color='blue')
plt.plot(time, pna, color='red')
plt.title('PNa and PK vs Time')
plt.xlabel('Time')
plt.ylabel('PNa and PK')
plt.grid(True)
plt.show()
```
